# APPLY method on columns

```python
# Define the function to be applied on columns
def last_four(num):
    return str(num)[-4:]
# Apply the function to selected column
# NOTE: FUNCTION should return a single value per calculation not lists or vectors
df['last_four']=df['CC Number'].apply(last_four)
df
```

```python
# Define the function
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
# apply the function to the selected column
df['yelp'] = df['total_bill'].apply(yelp)
df
```

# APPLY with lambda function on multiple columns

```python
# Define the function
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
# apply the lambda function to the columns
df['Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']), axis=1)
df
```

# APPLY with VECTORIZE

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
df
```

# MEASURING TIME EXECUTION WITH TIMEIT

```python
import timeit
# Define the setup statement
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv("C:/Users/AndresFelipeGonzalez/Documents/Udemy_capacitaciones/Python_ML_DS/UNZIP_FOR_NOTEBOOKS_FINAL/DATA/tips.csv")
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
'''
```

```python
# Define the first statement
stmt_one = '''
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
'''
# Define the first statement
stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
'''
```

```python
# configure the timeit with setup and first statement, then define the number of repetitions
timeit.timeit(setup=setup, stmt=stmt_one, number=1000)
```

```python
timeit.timeit(setup=setup, stmt=stmt_two, number=1000)
```

# CONCATENATION

```python
# Declare new datframes to concatenate
data_one = {'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']}
data_two = {'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']}
one = pd.DataFrame(data_one)
two = pd.DataFrame(data_two)
```

```python
# concatenate along the columns
pd.concat([one,two],axis=1)
```

```python
# Concatenate along the rows without matching the column names
pd.concat([one,two],axis=0)
```

```python
# Concatenate along the rows matching the column names
two.columns = one.columns
```

```python
pd.concat([one,two],axis=0)
```
